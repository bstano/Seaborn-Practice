```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from matplotlib import pyplot
```

We can use data from the World Happiness Report to try out some visualization methods.

```python
happiness = pd.read_csv('C:\Code\Data\happy2016.csv')

display(happiness.head(5))
```

Choose one variable and plot that variable four different ways.

"Happiness Score"

```python
plt.plot(happiness["Happiness Score"])
plt.title("Happiness Score By Country Rank")
plt.xlabel("Country Rank")
plt.ylabel("Happiness Score Value")
plt.show()
```

```python
plt.hist(happiness["Happiness Score"], )
plt.title("Global Happiness Score")
plt.xlabel("Happiness Score Value")
plt.show()
```

```python
sns.distplot(happiness["Happiness Score"])
plt.title("Global Happiness Score")
plt.xlabel("Happiness Score Value")
plt.show()
```

```python
plt.boxplot(happiness["Happiness Score"])
plt.title("Global Happiness Score Description")
plt.xticks([])
plt.ylabel("Happiness Score Value")
plt.show()
```

Choose two continuous variables, and plot them three different ways.

GDP
Life Expectancy

```python
g = sns.lmplot(y='Health (Life Expectancy)',
               x='Economy (GDP per Capita)',
               data=happiness,
               fit_reg=False,
               scatter_kws={'alpha':0.5})
g.set_ylabels("Economy Happiness Factor")
g.set_xlabels("GDP Happiness Factor")
plt.title('Life Expectancy by GDP')
plt.show()

g = sns.lmplot(y='Health (Life Expectancy)',
               x='Economy (GDP per Capita)',
               data=happiness,
               fit_reg=True,
               scatter_kws={'alpha':0.5})
g.set_ylabels("Economy Happiness Factor")
g.set_xlabels("GDP Happiness Factor")
plt.title('Life Expectancy by GDP')
plt.show()
```

```python
sns.set(style="darkgrid", color_codes=True)

g = sns.jointplot('Health (Life Expectancy)', 'Economy (GDP per Capita)', data=happiness, kind="reg", color="r", size=7)
```

Choose one continuous variable and one categorical variable, and plot them six different ways.

Happiness vs GDP score

```python
fig, ax = pyplot.subplots()
ax = sns.boxplot(y='Region',x='Happiness Score',data=happiness)
fig.set_size_inches(14, 8)
plt.suptitle('Happiness Score by Region')
sns.despine(offset=10, trim=True)
ax.set(ylabel='', xlabel='Happiness Score')


plt.show()
```

```python
fig, ax = pyplot.subplots()
ax = sns.boxplot(x='Region',y='Happiness Score',data=happiness,orient='v')
fig.set_size_inches(12, 8)
plt.suptitle('Happiness Score by Region')
sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Happiness Score')
ax.set_xticklabels(labels=['Western Europe','North America','Austrailia','Middle East and North Afica','Latin America',
                           'Southeast Asia','Central and Eastern Europe','Eastern Asia','Sub-Saharan Africa','Southern Asia'],
                   rotation=30)

plt.show()
```

```python
fig, ax = pyplot.subplots()
ax = sns.violinplot(y='Region', x='Happiness Score', data=happiness)
fig.set_size_inches(12, 7)
ax.set(ylabel='', xlabel='Happiness Score')
plt.show()
```

```python
sns.set(style="darkgrid")
g = sns.factorplot(x='Region', y='Happiness Score', data=happiness,
                   size=7, kind="bar", palette="pastel", ci=95)

g.despine(left=True)
g.set_ylabels("Happiness Score")
g.set_xlabels("")
g.set_xticklabels(rotation=70)
plt.title('Happiness Score by Area')
plt.show()
```

```python
fig, ax = pyplot.subplots()
ax = sns.barplot(y=happiness['Region'], x=happiness['Happiness Score'], palette="RdBu_r")
fig.set_size_inches(10, 7)
ax.set(ylabel='', xlabel='Happiness Score')
plt.suptitle('Happiness Score vs. Region')
plt.show()
```

```python
sns.set(style="whitegrid")

g = sns.factorplot(x='Region', y='Happiness Score', data=happiness,
                   size=7, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels('Happiness Score')
g.set_xlabels("")
g.set_xticklabels(rotation=70)
plt.title('Happiness Score by Area')
plt.show()
```

```python
sns.lmplot(y='Happiness Score',
           x='Economy (GDP per Capita)',
           data= happiness,
           fit_reg=False,
           col='Region',
           hue='Region',
           col_wrap=4,
           scatter_kws={'alpha':0.5})
plt.suptitle('Happiness Score by GDP Score')
plt.subplots_adjust(top=0.9)
plt.show()

sns.lmplot(y='Happiness Score',
           x='Economy (GDP per Capita)',
           data= happiness,
           fit_reg=True,
           col='Region',
           hue='Region',
           col_wrap=4,
           scatter_kws={'alpha':0.5})
plt.suptitle('Happiness Score by GDP Score')
plt.subplots_adjust(top=0.9)
plt.show()
```
